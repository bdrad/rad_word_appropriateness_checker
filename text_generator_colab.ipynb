```
def merge_files(files_path, outfile_path = "tiny_raw_text.txt"):
  '''
    delelt blank lines in each file and merge into one file
  '''
  with open(outfile_path, "w") as outfile:
    for f in files_path:
      with open(f, encoding = "utf-8") as infile:
        content = infile.read()
        for c in content:
          if c == '\n':
            continue
          outfile.write(c)
      outfile.write('\n\n')
```

```
def sent_seg(infile_path = "tiny_raw_text.txt", outfile_path = "tiny_text_9.txt"):
  '''
    Sentence segmentation on the input file. 
    Input file: raw documents delimited by empty lines.
    Output file: one sentence per line, documents are delimited by empty lines.

  '''

  nlp = English()  # load language
  sentencizer = nlp.create_pipe("sentencizer")
  nlp.add_pipe(sentencizer)

  with open(infile_path, "r") as infile, open(outfile_path, "w") as outfile:
    #print(infile)
    text = str()
    while True:
      # read each document seperately
      line = infile.readline()
      if not line: 
        break
      line_text = line.replace('\n','')
      if line_text:
        text += line_text.strip() +" "
        continue
      # perform sentence segmentation on each document
      doc = nlp(text)
      for sent in doc.sents:
        print(sent.text.strip())
        outfile.write(sent.text.strip()+'\n') #sent.text
      print('\n')
      outfile.write('\r\n') 
      text = ""
```
